```python
import pandas as pd
import plotly.express as px
```

```python
f = "../data/CNI_CNI20260406224057.csv"
# Load the data, using header=12 as it aligns with the start of numeric data
wide = pd.read_csv(f, header=12)
# Rename the columns to the desired descriptive names
wide = wide.rename(columns={
    wide.columns[0]: "fecha",
    wide.columns[2]: "Tonelas CO2"
})

# Convert relevant columns to numeric types for plotting
wide["fecha"] = pd.to_numeric(wide["fecha"])
wide["Tonelas CO2"] = pd.to_numeric(wide["Tonelas CO2"])

# Create the line plot with the correct column names
fig = px.line(wide, x="fecha", y="Tonelas CO2", title='Emision de CO2 por quema de combustibles fosiles')
fig.show()
```

```python

g = "../data/sector_sh_es.csv"
df = pd.read_csv(g)

# Rename the columns to the desired descriptive names
df = df.rename(columns={
    df.columns[0]: "Sector",
    df.columns[1]: "Periodo",
    df.columns[2]: "Indice de emisión"
})
# Convert relevant columns to numeric types for plotting
df["Periodo"] = pd.to_numeric(df["Periodo"])
df["Indice de emisión"] = pd.to_numeric(df["Indice de emisión"].str.replace(',', ''))
fig = px.line(df, x="Periodo", y="Indice de emisión", color='Sector')
fig.show()
```

```python

e = "../data/CNI_CNI20260405134428.csv"
# Load the data, using header=12 as it aligns with the start of numeric data
wid = pd.read_csv(e, header=12)
# Rename the columns to the desired descriptive names
wid = wid.rename(columns={
    wid.columns[0]: "Periodo",
    wid.columns[2]: "% de Hidrocarburos"
})

# Convert relevant columns to numeric types for plotting
wid["Periodo"] = pd.to_numeric(wid["Periodo"])
wid["% de Hidrocarburos"] = pd.to_numeric(wid["% de Hidrocarburos"])

# Create the line plot with the correct column names
fig = px.line(wid, x="Periodo", y="% de Hidrocarburos", title='Participacion de los hidrocarburos a nivel nacional')
fig.show()
```
