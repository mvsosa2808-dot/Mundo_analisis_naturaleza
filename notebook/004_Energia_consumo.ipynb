```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

data = """año	Población que cuenta con servicio de energía eléctrica (%)
2005	96.23
2006	96.73
2007	96.56
2008	96.72
2009	96.84
2010	97.6
2011	97.89
2012	98.11
2013	98.23
2014	98.43
2015	98.53
2016	98.58
2017	98.64
2018	98.68
"""

df = pd.read_csv(io.StringIO(data), sep='\t')

# Rename the second column for easier access
df.columns = ['año', 'porcentaje_energia_electrica']

df['año'] = df['año'].astype(int)

display(df.head())
plt.figure(figsize=(12, 6))
sns.lineplot(x='año', y='porcentaje_energia_electrica', data=df, marker='o')

plt.title('Población con Servicio de Energía Eléctrica (%) por Año')
plt.xlabel('Año')
plt.ylabel('Porcentaje de Población (%)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(df['año'], rotation=45)
plt.ylim(95, 100) # Set y-axis limits for better visualization of changes
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import plotly.express as px


g = "../data/Demanda_real_balance.csv"
df = pd.read_csv(g)

# Rename the columns to the desired descriptive names
df = df.rename(columns={
    df.columns[1]: "Area",
    df.columns[2]: "Hora",
    df.columns[3]: "Generacion (MWh)",
    df.columns[4]: "Importacion Total (MWh)",
    df.columns[5]: "Exportacion Total (MWh)",
    df.columns[7]: "Estimacion de Demanda por Balance (MWh)",
})
# Convert relevant columns to numeric types for plotting
df["Hora"] = pd.to_numeric(df["Hora"])
df["Generacion (MWh)"] = pd.to_numeric(df["Generacion (MWh)"])
fig = px.line(df, x="Hora", y="Generacion (MWh)", color='Area')
fig.show()

# Convert relevant columns to numeric types for plotting
df["Hora"] = pd.to_numeric(df["Hora"])
df["Importacion Total (MWh)"] = pd.to_numeric(df["Importacion Total (MWh)"])
fig = px.line(df, x="Hora", y="Importacion Total (MWh)", color='Area')
fig.show()

# Convert relevant columns to numeric types for plotting
df["Hora"] = pd.to_numeric(df["Hora"])
df["Exportacion Total (MWh)"] = pd.to_numeric(df["Exportacion Total (MWh)"])
fig = px.line(df, x="Hora", y="Exportacion Total (MWh)", color='Area')
fig.show()

# Convert relevant columns to numeric types for plotting
df["Hora"] = pd.to_numeric(df["Hora"])
df["Estimacion de Demanda por Balance (MWh)"] = pd.to_numeric(df["Estimacion de Demanda por Balance (MWh)"])
fig = px.line(df, x="Hora", y="Estimacion de Demanda por Balance (MWh)", color='Area')
fig.show()
```
